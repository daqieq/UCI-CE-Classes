# Plot Activation Functions and Their Derivatives

## Identity

```python
import numpy as np
import matplotlib.pyplot as plt

def identity_activation(x):
    return x

x = np.array(np.arange(-10,10,0.1))
y = identity_activation(x)
dydx = np.ones_like(x)

%matplotlib inline
plt.plot(x, y, 'b', label='Identity')
plt.plot(x, dydx, 'r', label='Derivative', linestyle='--')
#plt.axhline(y=1, color='r', label='Derivative', linestyle='--')
plt.legend(loc='upper center', shadow=True)
plt.show()
```

## TanH

```python
def tanh_activation(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

x = np.array(np.arange(-10,10,0.1))
y = tanh_activation(x)
dydx = 1-np.square(y)

%matplotlib inline
plt.plot(x, y, 'b', label='TanH')
plt.plot(x, dydx, 'r', label='Derivative', linestyle='--')
plt.legend(loc='lower right', shadow=True)
plt.show()
```

## ReLU

```python
def relu_activation(x):
    return np.where(x >= 0, x, 0)

x = np.array(np.arange(-10,10,0.1))
y = relu_activation(x)
dydx = np.where(x >= 0, 1, 0)

%matplotlib inline
plt.plot(x, y, 'b', label='ReLU')
plt.plot(x, dydx, 'r', label='Derivative', linestyle='--')
plt.legend(loc='upper left', shadow=True)
plt.show()
```
